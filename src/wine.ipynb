```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from main import ConvexHull
```

```python
wine = datasets.load_wine()

#create a DataFrame Wine
df1 = pd.DataFrame(wine.data, columns=wine.feature_names)
df1['Target'] = pd.DataFrame(wine.target)
print(df1.head())
```

```python
#visualisasi hasil ConvexHull

plt.figure(figsize = (10, 6))
colors = ['b','r','g']
plt.title('Alcohol and Color Intensity')
plt.xlabel(wine.feature_names[0])
plt.ylabel(wine.feature_names[9])
for i in range(len(wine.target_names)):
    bucket = df1[df1['Target'] == i]
    bucket = bucket.iloc[:,[0,9]].values
    edges = ConvexHull(bucket)
    plt.scatter(bucket[:, 0], bucket[:, 1], label=wine.target_names[i])
    for j in range(len(edges)):
        plt.plot(bucket[edges[j], 0], bucket[edges[j], 1], colors[i])
plt.legend()
```
